```python
import pandas as pd
!pip install pycaret
from pycaret.classification import load_model, predict_model

def predict_churn(data,thres=0.7):
  model = load_model('/content/drive/MyDrive/Colab Notebooks/final_gradient_boost_classifier1')
  predictions = predict_model(model, data=data)
  label_col = 'Label' if 'Label' in predictions.columns else 'prediction_label'
  score_col = 'Score' if 'Score' in predictions.columns else 'prediction_score'
  predictions['churn_prediction']=predictions['prediction_score']>thres
  predictions['churn_prediction'].replace({True:0,False:1},inplace=True)
  return predictions[['churn_prediction']]


if __name__ == "__main__":

    new_data = pd.read_csv('/content/drive/MyDrive/new_churn_data.csv',index_col="customerID")
    probabilities = predict_churn(new_data)
    print(probabilities)
```
